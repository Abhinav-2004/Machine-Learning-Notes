# Polynomial Regression

## Importing the libraries

```
import pandas as pd
import matplotlib.pyplot as mlt
import numpy as np
```

## Importing the dataset

```
dataset=pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

## Training the Linear Regression model on the whole dataset

```

from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

## Training the Polynomial Regression model on the whole dataset

```
#polynomial regression model
#y=b0 + b1x1 + b2(x1)^2 + b3(x1)^3......+ bn(x1)^n
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4) #degree is the value of n
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualising the Linear Regression results

```
mlt.scatter(X, y,color='red')
mlt.plot(X,lin_reg.predict(X), color='blue')
mlt.title("Truth or Bluff")
mlt.xlabel("Position Label")
mlt.ylabel("Salary")
mlt.show()

#red points are real points and model is clearly unfit
```

## Visualising the Polynomial Regression results

```
mlt.scatter(X, y, color='red')
mlt.plot(X, lin_reg_2.predict(X_poly), color='blue')
mlt.title("Truth or Bluff")
mlt.xlabel("Position Label")
mlt.ylabel("Salary")
mlt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
mlt.scatter(X, y, color = 'red')
mlt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
mlt.title('Truth or Bluff (Polynomial Regression)')
mlt.xlabel('Position level')
mlt.ylabel('Salary')
mlt.show()
```

## Predicting a new result with Linear Regression

```
lin_reg. predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
lin_reg_2.predict(poly_reg.fit_transform([[6.5]])) #x1=6.5 since power is 4 so input it accordingly as x1, x1^2 ,x1^3...
```
